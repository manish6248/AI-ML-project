# Split Dataset into Training set and Test set

```python
#split dataset into features and label
X= df2.drop(labels='left',axis=1)
y= df2['left']
```

```python
#split dataset into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
len(X_train)
```

```python
len(X_test)
```

# Model Selection

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
```

```python
model_param = {

    'RandomForestClassifier':{
        'model': RandomForestClassifier(),
        'param':{
            'n_estimators': [100, 200],              # more stable
            'criterion': ['gini', 'entropy'],
            'max_depth': [5, 10, None],              # was too small
            'max_features': ['sqrt', 'log2']
        }
    },

    'XGBClassifier':{
        'model': XGBClassifier(objective='binary:logistic'),
        'param':{
            'learning_rate': [0.05, 0.1, 0.2],       # removed 0.5, 0.001
            'max_depth': [3, 5, 7],                  # removed 10, 20
            'n_estimators': [100, 200, 300],         # more trees
            'subsample': [0.8, 1.0],                 # added (important)
            'colsample_bytree': [0.8, 1.0]           # added (important)
        }
    }
}
```

```python
scores = []

for model_name, mp in model_param.items():
    model_selection = GridSearchCV(
        estimator=mp['model'],
        param_grid=mp['param'],
        cv=5,
        return_train_score=False
    )


    model_selection.fit(X_train, y_train)

    scores.append({
        'model': model_name,
        'best_score': model_selection.best_score_,
        'best_params': model_selection.best_params_
    })

```

```python
scores
```

```python
# as per above results, xgboost gives best result

model_xgb = XGBClassifier(
    objective='binary:logistic',
    learning_rate=0.1,
    max_depth=5,
    n_estimators=200
)

model_xgb.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracies = []

for i in range(10):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=i
    )
    
    model_xgb.fit(X_train, y_train)
    y_pred = model_xgb.predict(X_test)
    
    accuracies.append(accuracy_score(y_test, y_pred))

print("All Accuracies:", accuracies)
print("Average Accuracy:", sum(accuracies)/len(accuracies))
```

# Cross-validation

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(model_xgb, X, y, cv=5, scoring='accuracy')

print("CV Scores:", cv_scores)
print("Mean CV Accuracy:", cv_scores.mean())
```

# Evaluation (Train vs Test)

```python
# Train accuracy
train_acc = model_xgb.score(X_train, y_train)

# Test accuracy
test_acc = model_xgb.score(X_test, y_test)

print("Train Accuracy:", train_acc)
print("Test Accuracy:", test_acc)
```

# F1 + ROC-AUC

```python
from sklearn.metrics import f1_score, roc_auc_score

y_pred = model_xgb.predict(X_test)
y_prob = model_xgb.predict_proba(X_test)[:,1]

print("F1 Score:", f1_score(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```

# Confusion Matrix

```python

from sklearn.metrics import confusion_matrix
import seaborn as sn
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
